# 구글스토어에서 라인버블2 리뷰 크롤링(셀레니움)

```python
import time
import requests
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException 
from selenium.webdriver.common.keys import Keys #스크롤다운의 key
```

```python
driver = webdriver.Chrome('C:/Users/snari/Downloads/chromedriver_win32/chromedriver.exe')
url = "https://play.google.com/store/apps/details?id=com.linecorp.LGBB2&showAllReviews=true"
driver.get(url)

#페이지 로딩
# By.ID - 태그에 있는 ID 로 검색
# By.CSS_SELECTOR - CSS Selector 로 검색
# By.NAME - 태그에 있는 name 으로 검색
# By.TAG_NAME - 태그 이름으로 검색
try:
    myElem = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, '#fcxH9b > div.WpDbMd > c-wiz > div > div.ZfcPIb > div > div.JNury.Ekdcne > div > div > div.W4P4ne > div:nth-child(2) > div:nth-child(3) > div:nth-child(1) > div > div.d15Mdf.bAhLNe > div.UD7Dzf')))
    print("크롤링 준비 완료")
except TimeoutException:
    print("페이지 로딩 시간 초과")
    exit()
except Exception as e:
    print(e)
    exit()
```

```python
def check_end():
    try:
        message = driver.find_element_by_css_selector('#fcxH9b > div.WpDbMd > c-wiz > div > div.ZfcPIb > div > div.JNury.Ekdcne > div > div > div.W4P4ne > div:nth-child(2) > div.PFAhAf > div > span > span')
    except NoSuchElementException:
        return False
    else: return True
```

```python
#html body 영역
elem = driver.find_element_by_tag_name("body")
```

```python
print("스크롤 다운 시작")
l = 0
while not check_end(): #check_end가 TRUE가 리턴될때까지 실행
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
    l += 1
    if l % 5 == 0:
        print(l, "번째 스크롤 다운")
    # 6번째서 종료
    #     if l == 6:
    #         break
    try:
        driver.find_element_by_xpath('//*[@id="fcxH9b"]/div[4]/c-wiz/div/div[2]/div/div[1]/div/div/div[1]/div[2]/div[2]/div/span/span').click()
    except:
        None
print("스크롤 다운 종료")
```

```python
#스크롤 다운 후 크롤링
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
review_text = []
star_text = []
day_text = []
review_id_text = []
good_text = []
result = {}

review = soup.select('div.UD7Dzf > span:nth-child(1)')
star = soup.select('div.bAhLNe.kx8XBd > div > span.nt2C1d > div > div')
day = soup.select('span.p2TkOb')
review_id = soup.select('div.bAhLNe.kx8XBd > span.X43Kjb')
good = soup.select('div.jUL89d.y92BAb')

for index in range(len(review_id)):
    review_text.append(review[index].text.replace(' ...전체 리뷰',''))
    star_text.append(int(star[index]['aria-label'].replace('별표 5개 만점에 ','').replace('개를 받았습니다.','')))
    day_text.append(day[index].text)
    review_id_text.append(review_id[index].text)
    good_text.append(good[index].text)

    
result = {'아이디':review_id_text,'날짜':day_text,'좋아요':good_text,'별점':star_text,'review':review_text}
df = pd.DataFrame(result)
```
